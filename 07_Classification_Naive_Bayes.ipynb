```r
# 01. SETUP E REAPROVEITAMENTO DE DADOS

# Carregar bibliotecas (se não estiverem carregadas):
library(caret)
library(mlbench) 
library(dplyr)
# ... (Código para carregar df, renomear diabetes, set.seed(42)) ...

# Reutilizar o conjunto de treino e o controle
# train_df, test_df, control
```

```r
# Carregar dados
load("data/03_classification_data_splits.RData") 

# Carregar as configurações de controle
control <- readRDS("config/train_control.rds")

# Carregar o modelo RegLog Otimizado para comparação
logreg_otimizado <- readRDS("models/logreg_otimizado.rds")

# Verificar se carregou
print(dim(train_df))
```

```r
# 03. TREINAMENTO DO MODELO NAIVE BAYES (NB)
# Instalar o pacote necessário para o Naive Bayes (NB)
install.packages("klaR")

# Carregar o pacote
library(klaR)

message("Treinando Modelo Naive Bayes (NB)...")

# O Naive Bayes é um modelo de ajuste (fit) e não requer otimização de parâmetros
# na sua forma mais básica (método 'nb').

nb_model <- train(
    diabetes ~ ., 
    data = train_df, 
    method = "nb",                   # Método para Naive Bayes
    metric = "ROC", 
    trControl = control, 
    preProcess = c("center", "scale")
)

# Resumo da performance na Validação Cruzada
print(nb_model)
```

```r
# Salvar o modelo final (muito útil para a comparação)
saveRDS(logreg_otimizado, file = "models/nb_model.rds")
```
