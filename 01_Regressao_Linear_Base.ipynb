```r
# Code setup

library(tidyverse)


data(mtcars)


set.seed(42)
```

```r
# Visualização da relação entre o alvo (mpg) e um feature (wt)


mtcars %>%


ggplot(aes(x = wt, y = mpg)) +


geom_point(color = "#0072B2") +


geom_smooth(method = "lm", se = FALSE) +


labs(title = "Relação Linear entre MPG e Peso")
```

```r
## Regressão Linear Simples: mpg ~ wt

modelo_simples <- lm(mpg ~ wt, data = mtcars)

summary(modelo_simples)
```

```r
# Diagnóstico do Modelo
# Checa as suposições do modelo: linearidade, normalidade dos 
# resíduos, homocedasticidade.

par(mfrow = c(2, 2))


plot(modelo_simples)

```

```r
## Regressão Linear Múltipla: mpg ~ wt + hp

modelo_multiplo <- lm(mpg ~ wt + hp, data = mtcars)

summary(modelo_multiplo)
```

```r
# Exemplo de Previsão
# Novo dado: Peso = 3.000 lbs, Horsepower = 150

novo_dado <- data.frame(wt = 3.00, hp = 150)


previsao_mpg <- predict(modelo_multiplo, novo_dado)


cat("Previsão de MPG para o novo veículo:", round(previsao_mpg, 2))
```
