```r
# --- CÉLULA 3: TREINAMENTO DO MODELO SVM (KERNEL RBF) ---

# O método 'svmRadial' no caret requer o pacote 'kernlab'
library(kernlab) 

message("Treinando Modelo Support Vector Machines (SVM)...")

# O SVM otimiza dois hiperparâmetros:
# 1. C (Cost): Penalidade por erro de classificação (soft margin)
# 2. sigma: Parâmetro do kernel RBF, que define o alcance da influência de um único ponto de treino.

svm_model <- train(
    diabetes ~ ., 
    data = train_df, 
    method = "svmRadial",           # SVM com Kernel Radial
    metric = "ROC", 
    trControl = control, 
    preProcess = c("center", "scale"), # Padronização é CRUCIAL para SVM
    tuneLength = 9                  # O caret cria uma grade de 9 valores para C e sigma
)


# Salvar o modelo final (muito útil para a comparação)
saveRDS(svm_model, file = "models/svm_model.rds")

print(svm_model)
```
