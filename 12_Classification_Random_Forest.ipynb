```r
# --- CÉLULA 3: TREINAMENTO DO MODELO RANDOM FOREST (RF) ---

message("Treinando Modelo Random Forest...")

# Otimiza o parâmetro 'mtry' (número de variáveis a serem amostradas)
rf_model <- train(
    diabetes ~ ., 
    data = train_df, 
    method = "rf",                   # Método para Random Forest
    metric = "ROC", 
    trControl = control, 
    # O Random Forest é robusto à padronização, mas mantemos o controle.
    tuneLength = 8,                  # O caret testará 8 valores de mtry
    ntree = 500                      # Número de árvores (500 é um bom padrão)
)

# Salvar o modelo final (muito útil para a comparação)
saveRDS(rf_model, file = "models/rf_model.rds")

print(rf_model)
```
