```r
# --- CÉLULA 3: TREINAMENTO DO MODELO GRADIENT BOOSTING MACHINES (GBM) ---

message("Treinando Modelo Gradient Boosting Machines (GBM)...")

# O GBM tende a ser mais lento devido à otimização sequencial
gbm_model <- train(
    diabetes ~ ., 
    data = train_df, 
    method = "gbm",                   # Método para Gradient Boosting Machines
    metric = "ROC", 
    trControl = control, 
    verbose = FALSE,                  # Evita a impressão de logs detalhados
    tuneGrid = expand.grid(
        n.trees = c(50, 100, 150),      # Número de árvores
        interaction.depth = c(1, 2, 3), # Profundidade de interação
        shrinkage = c(0.01, 0.1),       # Taxa de aprendizado
        n.minobsinnode = 10             # Mínimo de observações por nó
    )
)

# Salvar o modelo final (muito útil para a comparação)
saveRDS(gbm_model, file = "models/gbm_model.rds")

print(gbm_model)
```
