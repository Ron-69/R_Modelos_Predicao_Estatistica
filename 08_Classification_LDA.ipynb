```r
# 01. SETUP E REAPROVEITAMENTO DE DADOS

# Carregar bibliotecas (se não estiverem carregadas):
library(caret)
library(mlbench) 
library(dplyr)
# ... (Código para carregar df, renomear diabetes, set.seed(42)) ...
set.seed(42)
# Reutilizar o conjunto de treino e o controle
# train_df, test_df, control
```

```r
# 02. Carregar dados
load("data/03_classification_data_splits.RData") 

# Carregar as configurações de controle
control <- readRDS("config/train_control.rds")

# Carregar o modelo RegLog Otimizado para comparação
logreg_otimizado <- readRDS("models/logreg_otimizado.rds")

# Carregar o modelo RegLog Otimizado para comparação
nb_model <- readRDS("models/nb_model.rds")

# Verificar se carregou
print(dim(train_df))
```

```r
# 03. TREINAMENTO DO MODELO LINEAR DISCRIMINANT ANALYSIS (LDA)

message("Treinando Modelo Linear Discriminant Analysis (LDA)...")

# Reutilizando o train_df e o control
lda_model <- train(
    diabetes ~ ., 
    data = train_df, 
    method = "lda",                  # Método para Linear Discriminant Analysis
    metric = "ROC", 
    trControl = control, 
    preProcess = c("center", "scale")
)

# Salvar o modelo final (muito útil para a comparação)
saveRDS(lda_model, file = "models/lda_model.rds")

print(lda_model)
```

```r
# 05. COMPARAÇÃO DOS MODELOS PROBABILÍSTICOS (RegLog vs. NB vs. LDA)

# Lembre-se de carregar o logreg_otimizado (se estiver em um novo notebook)
# logreg_otimizado <- readRDS("models/logreg_otimizado.rds") 

message("Comparação de Desempenho dos 3 Modelos Probabilísticos:")

prob_models_comp <- resamples(list(
    RegLog_Opt = logreg_otimizado, # Modelo Otimizado (da seção anterior)
    Naive_Bayes = nb_model,        # Modelo recém-treinado
    LDA = lda_model                # Modelo recém-treinado
))

print(summary(prob_models_comp))
```
